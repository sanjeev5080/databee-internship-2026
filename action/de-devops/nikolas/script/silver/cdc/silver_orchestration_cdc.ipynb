## Orchestration Logic
This notebook programmatically runs all Silver transformation notebooks in sequence.  
It becomes the **single entry point** for the Silver layer in Databricks Jobs.

```python
notebooks = [
    "./crm/silver_crm_cust_info_cdc",
    "./crm/silver_crm_prd_info_cdc",
    "./crm/silver_crm_sales_details_cdc",
    "./erp/silver_erp_cust_cdc",
    "./erp/silver_erp_loc_cdc",
    "./erp/silver_erp_pxcat_cdc",
    "./api/silver_holidays_cdc"
]

for nb in notebooks:
    print(f"Running {nb}")
    try:
        result = dbutils.notebook.run(nb, timeout_seconds=0)
        print(f"  ✓ {result if result else 'Success'}")
    except Exception as e:
        error_msg = str(e)
        if "No new data" in error_msg:
            print(f"  ℹ️  No new data to process")
        else:
            print(f"  ✗ Failed: {error_msg}")
            raise
```
