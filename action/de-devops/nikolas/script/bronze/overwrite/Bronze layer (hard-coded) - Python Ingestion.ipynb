## Bronze Layer - Python Ingestion

This notebook ingests raw CSV files from Volumes into Bronze layer Delta tables using Python.

**Data Sources:**
- **CRM Source**: Customer info, product info, sales details
- **ERP Source**: Customer records, location data, product categories

All tables are created using `df.write.mode("overwrite").saveAsTable()`

```python
df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/bronze/raw_sources/source_crm/cust_info.csv")
)

df.display()
```

## Write it to Bronze layer

```python
df.write.mode("overwrite").saveAsTable("bronze.crm_cust_info")
```
