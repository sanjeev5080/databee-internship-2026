## Bronze Layer - SQL Ingestion

This notebook ingests raw CSV files from Volumes into Bronze layer Delta tables using SQL.

**Data Sources:**
- **CRM Source**: Customer info, product info, sales details
- **ERP Source**: Customer records, location data, product categories

All tables are created using `CREATE TABLE ... AS SELECT * FROM read_files()`

```python
%sql
CREATE OR REPLACE TABLE bronze.crm_cust_info
USING DELTA
AS
SELECT * FROM read_files('/Volumes/workspace/bronze/raw_sources/source_crm/cust_info.csv', format => 'csv', header => true);
```
