# Holiday Impact Analysis - Business Metrics

Analyzes sales performance during holidays vs non-holidays. Answers:
- Which holidays drive the most revenue?
- What's the average order value during holidays?
- How many unique customers shop on each holiday?
- Compare holiday vs non-holiday sales patterns

```python
%sql
CREATE OR REPLACE TABLE workspace.gold.agg_holiday_impact AS
SELECT 
  holiday_name,
  COUNT(DISTINCT order_number) AS orders_on_holiday,
  SUM(sales_amount) AS revenue_on_holiday,
  ROUND(AVG(sales_amount), 2) AS avg_order_value,
  COUNT(DISTINCT customer_key) AS unique_customers,
  SUM(quantity) AS total_items_sold

FROM workspace.gold.fact_sales_enriched
WHERE is_holiday = TRUE AND order_date IS NOT NULL
GROUP BY holiday_name
ORDER BY revenue_on_holiday DESC
```

```python
%sql
SELECT 
  holiday_name,
  orders_on_holiday,
  revenue_on_holiday,
  avg_order_value,
  unique_customers
FROM workspace.gold.agg_holiday_impact
ORDER BY revenue_on_holiday DESC
LIMIT 10
```

```python
%sql
SELECT 
  CASE WHEN is_holiday THEN 'Holiday' ELSE 'Non-Holiday' END AS period_type,
  COUNT(DISTINCT order_number) AS total_orders,
  SUM(sales_amount) AS total_revenue,
  ROUND(AVG(sales_amount), 2) AS avg_order_value,
  COUNT(DISTINCT customer_key) AS unique_customers
FROM workspace.gold.fact_sales_enriched
WHERE order_date IS NOT NULL
GROUP BY CASE WHEN is_holiday THEN 'Holiday' ELSE 'Non-Holiday' END
```
